```python
new_data = pd.read_csv("C:/Users/USER/Documents/QQQ Project/data/raw/new_QQQ_data.csv")
new_data = pd.DataFrame(new_data)
print(new_data.columns)
print(new_data.index)
```

```python
# Clean old data
old_data = old_data.drop(columns='Close')
old_data = old_data.rename(columns={'Adj Close':'Close'})
print(old_data.columns)
```

```python
# Clean new data 
new_data = new_data.rename(columns={'Price':'Date'})
new_data = new_data.set_index('Date')
print(new_data.columns)
```

```python
new_data = new_data[['Open', 'High', 'Low', 'Close', 'Volume']]
full_dataset = pd.concat([old_data,new_data])
full_dataset = full_dataset.sort_index()
full_dataset.to_csv("C:/Users/USER/Documents/QQQ Project/data/processed/full_QQQ_data.csv")
```
