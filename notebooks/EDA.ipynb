# Proper EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv("../data/processed/new_QQQ_data.csv",
                 parse_dates=["Date"],
                 index_col="Date")
```

```python
df.head()
```

```python
df.info()
df.describe()
```

1999 - Present Plot

```python
plt.figure(figsize=(12,5))
plt.plot(df["Close"])
plt.title("QQQ Close Price (1999Present)")
plt.grid(True)
plt.show()
```

```python
df["returns"] = df["Close"].pct_change()

plt.figure(figsize=(12,5))
plt.plot(df["returns"])
plt.title("Daily Returns")
plt.grid(True)
plt.show()
```

```python
df["rolling_vol"] = df["returns"].rolling(20).std()

plt.figure(figsize=(12,5))
plt.plot(df["rolling_vol"])
plt.title("20-Day Rolling Volatility")
plt.grid(True)
plt.show()
```
